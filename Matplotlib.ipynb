# Data Visualization

# Introduction

This notebook provides a brief introduction to data visualization tools and principles. We will explore basic concepts using the Matplotlib library in Python and practice fundamental design principles.

# 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 2. Generate and plot data:
- x = [1, 2, 3, 4, 5]
- y = [3, 5, 7, 2, 1]

### Plot the data as a scatter plot

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

plt.scatter(x, y, color='blue', marker='o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(False)
plt.show()
```

# 3. Customize plot elements: 

- Change the plot market and color
- Add grid lines 
- Add proper Labels and Title

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

plt.scatter(x, y, color='green', marker='^', label='Scatter Points')
plt.title('Scatter Plot of X and Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

# 4. Create different chart types: 
- Line Plot
- Bar Plot
- Histogram

```python
plt.plot(x, y, color='blue', marker='o', linestyle='-', label='Line Plot')
plt.title('Line Plot of X and Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.bar(x, y, color='orange', label='Bar Plot')
plt.title('Bar Plot of X and Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

plt.hist(data, bins=5, color='purple', edgecolor='black', alpha=0.7, label='Histogram')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

# 5. Enhance plot with annotations: 
- Add an annotation on a random point in plot

```python
import random

random_index = random.randint(0, len(x) - 1)
annotated_point = (x[random_index], y[random_index])

plt.plot(x, y, color='blue', marker='o', linestyle='-', label='Line Plot')
plt.scatter(*annotated_point, color='red', marker='*', s=200, label='Annotated Point')

plt.annotate(f'Annotation Point\n({annotated_point[0]}, {annotated_point[1]})',
             xy=annotated_point,
             xytext=(annotated_point[0] + 0.5, annotated_point[1] + 1),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             bbox=dict(boxstyle='round,pad=0.3', edgecolor='black', facecolor='lightyellow'))

plt.title('Line Plot with Annotation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```

# 6. Subplots and multiple plots: 
Create 2 Subplots
+ Bar Plot
+ Scatter Plot

```python
x = [1, 2, 3, 4, 5]
y_bar = [3, 5, 7, 2, 1]
y_scatter = [4, 2, 6, 1, 5]

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].bar(x, y_bar, color='orange', label='Bar Plot')
axs[0].set_title('Bar Plot')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')
axs[0].grid(axis='y', linestyle='--', alpha=0.7)
axs[0].legend()

axs[1].scatter(x, y_scatter, color='green', marker='o', label='Scatter Plot')
axs[1].set_title('Scatter Plot')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()

plt.show()
```

# 7. Saving plots: 
Save the most recent plot as PNG

```python
x = [1, 2, 3, 4, 5]
y_bar = [3, 5, 7, 2, 1]
y_scatter = [4, 2, 6, 1, 5]

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].bar(x, y_bar, color='orange', label='Bar Plot')
axs[0].set_title('Bar Plot')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')
axs[0].grid(axis='y', linestyle='--', alpha=0.7)
axs[0].legend()

axs[1].scatter(x, y_scatter, color='green', marker='o', label='Scatter Plot')
axs[1].set_title('Scatter Plot')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()

plt.savefig('combined_plot.png')

plt.show()
```

# 8. Color theory and application:

##### Explore Color Palattes
Assign new variables
- b = range(100)
- a = [i**2 for i in x]

Create a Scatter Plot with Viridis Colormap

```python
#x = [1, 2, 3, 4, 5]
b = range(100)
a = [i**2 for i in b]

plt.scatter(a, b, c=b, cmap='viridis', marker='o', edgecolors='black')

plt.colorbar(label='b')

plt.title('Scatter Plot with Viridis Colormap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

#### Apply Color Maps to Data
Generate a 10x10 heatmap of random data using the 'coolwarm' colormap and display the color bar with labeled data values.


```python
random_data = np.random.rand(10, 10)

heatmap = plt.imshow(random_data, cmap='coolwarm', interpolation='nearest')

plt.colorbar(heatmap, label='Data Values')

plt.title('Heatmap with Coolwarm Colormap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

# 9. Layout and composition:

Explore the impact of different plot sizes and aspect ratios by creating a small square plot and a larger wide plot using the 'scatter' function.
- Small = 5x5
- Large = 10x5

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.figure(figsize=(5, 5))
plt.scatter(x, y, color='blue', marker='o')
plt.title('Small Square Plot (5x5)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='green', marker='s')
plt.title('Large Wide Plot (10x5)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

Modify the legend placement to the upper left and adjust the positioning of the annotation for the point (2, 5) in a scatter plot

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', marker='o', label='Data Points')
plt.title('Scatter Plot with Legend and Annotation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend(loc='upper left')

annotated_point = (2, 5)
plt.scatter(*annotated_point, color='red', marker='*', s=200, label='Annotated Point (2, 5)')
plt.annotate(f'Annotation Point\n({annotated_point[0]}, {annotated_point[1]})',
             xy=annotated_point,
             xytext=(annotated_point[0] + 1, annotated_point[1] + 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             bbox=dict(boxstyle='round,pad=0.3', edgecolor='black', facecolor='lightyellow'))

plt.grid(True)
plt.legend()
plt.show()
```

Enhance the use of whitespace in a plot by adjusting the spacing around the plot area and creating a scatter plot with improved whitespace.

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.figure(figsize=(8, 5))

plt.scatter(x, y, color='blue', marker='o', label='Data Points')
plt.title('Scatter Plot with Improved Whitespace')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.margins(0.05)
plt.tight_layout(pad=2.0)

plt.grid(True)
plt.legend()
plt.show()
```

# 10. Effective chart selection: 

Choose appropriate chart type for the following data
- categories = ['A', 'B', 'C']
- values = [10, 20, 30]

```python
categories = ['A', 'B', 'C']
values = [10, 20, 30]

plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart for Categories A, B, and C')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

# 11. Customize plot styles: 

Experiment with line styles, marker shapes, and color gradients:
- Dashed Line
- Triangle Marker (Green)

Utilize color gradients to represent data using a specified colormap, for example, 'YlGnBu', to create a scatter plot with a color bar.

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x, y, linestyle='--', color='blue', marker='^', markersize=8, label='Data Line')

colors = np.arange(len(x))
scatter = plt.scatter(x, y, c=colors, cmap='YlGnBu', marker='o', edgecolors='black', alpha=0.8, label='Scatter Points')

plt.colorbar(scatter, label='Data Value')

plt.title('Customized Plot with Dashed Line, Triangle Marker, and Color Gradients')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()
plt.grid(True)
plt.show()
```
